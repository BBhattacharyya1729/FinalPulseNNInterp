```julia
### This file does a Givens Rotation gate (1 parameter, 2 qubits) with X,Y, and ZZ control
using QuantumCollocation
using NamedTrajectories
using TrajectoryIndexingUtils
```

```julia
const Units = 1e9
const MHz = 1e6 / Units
const GHz = 1e9 / Units
const ns = 1e-9 * Units
const μs = 1e-6 * Units
;
```

```julia
# Operators
const Paulis = Dict(
    "I" => Matrix{ComplexF64}([1 0; 0 1]),
    "X" => Matrix{ComplexF64}([0 1; 1 0]),
    "Y" => Matrix{ComplexF64}([0 im; -im 0]),
    "Z" => Matrix{ComplexF64}([1 0; 0 -1]),
)
const a = [0 1; 0 0]
const ad = transpose(a);

H_drift = [ ]

excitation(theta) = exp(-im/2 * theta * (indexedOp("X",1)*indexedOp("Y",2)-indexedOp("Y",1)*indexedOp("X",2)));

function indexedOp(str, i)
    init = [Paulis["I"] for j in 1:2]
    init[i] = Paulis[str]
    return reduce(kron,init)
end


X_controls = [indexedOp("X", i) for i in 1:2]
Y_controls = [indexedOp("Y", i) for i in 1:2]
XY_controls = [(indexedOp("X",1)*indexedOp("X",2)+indexedOp("Y",1)*indexedOp("Y",2))/2]

H_drives = reduce(vcat,[X_controls,Y_controls,XY_controls])

system = QuantumSystem(H_drives);
t_f = 10 * ns
n_steps = 51
times = range(0, t_f, n_steps)  # Alternative: collect(0:Δt:t_f)
Δt = times[2] - times[1]
n_controls=5
n_qubits=2;
```

```julia
### Generate Initial Trajectories 
PICO_max_iter = 100

# Shape the cost function with weights on states and controls
Q = 1000.
R =1e-2
R_b=200.0
# Add control bounds
a_bound= 1.0
dda_bound = 1.0

ops = Options()
ops.print_info_string = "yes"
ops.recalc_y = "yes"
ops.recalc_y_feas_tol = 1.0 ##down
ops.print_level = 2
p = UnitarySmoothPulseProblem(
                system,
                excitation(pi),
                n_steps,
                Δt;
                a_bound=a_bound,
                dda_bound=dda_bound,
                Q=1000.0,
                R=1e-2,
                verbose=true,
                hessian_approximation=true,
                pade_order=10,
                free_time=true,
                timesteps_all_equal=true,
                max_iter=PICO_max_iter,
                ipopt_options=ops,
            )
min_time_problem  = UnitaryMinimumTimeProblem(p; final_fidelity=1-1e-5,max_iter=100)
solve!(min_time_problem)
```

```julia
Δt = ceil(1.2 * min_time_problem.trajectory[:Δt][1]*100)/100
```

```julia
N = 11


trajectory_list,glued_trajectory_list = Interpolator1D(
    excitation,
    system,
    n_steps,
    N,
    Δt;
    Q=Q,
    R=R,
    R_b=R_b,
    a_bound=a_bound,
    dda_bound=dda_bound,
    ipopt_options=ops,
    max_iter=PICO_max_iter
);
```

```julia
DATA = sample1D(trajectory_list,Δt,n_qubits,system,2500,N,excitation)
findmax(DATA)
```
